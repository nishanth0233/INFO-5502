# Task 3 (Identify the missing values and impute them).

```python
plt.figure(figsize=(8,6),dpi=100)
df.isnull().sum().plot(kind='bar')
```

```python
df.isnull().sum()
```

```python
df['TotalCharges']=df['TotalCharges'].fillna(df['TotalCharges'].mean())
```

```python
df.isnull().sum()
```

```python
df.to_csv('visualization.csv',index=False)
```

# task 4 encode the categorical values to binary integers.

```python
for i in df.columns:
    if df[i].dtypes=='O':
        print(df[i].value_counts())
```

```python
df['StreamingTV']=df['StreamingTV'].replace({'No internet service':'No'})
df['StreamingMovies']=df['StreamingMovies'].replace({'No internet service':'No'})
df['TechSupport']=df['TechSupport'].replace({'No internet service':'No'})
df['DeviceProtection']=df['DeviceProtection'].replace({'No internet service':'No'})
df['OnlineSecurity']=df['OnlineSecurity'].replace({'No internet service':'No'})
df['MultipleLines']=df['MultipleLines'].replace({'No phone service':'No'})
df['OnlineBackup']=df['OnlineBackup'].replace({'No internet service':'No'})
```

```python
for i in df.columns:
    if df[i].dtypes=='O':
        print(df[i].value_counts())
```

```python
df.columns
```

```python
value_list1=[]
for i in df.columns:
    if (len(df[i].value_counts())<=2) & (df[i].dtypes=='O'):
        print(i+':',len(df[i].value_counts()))
        value_list1.append(i)
value_list1
```

```python
value_list2=[]
for i in df.columns:
    if (len(df[i].value_counts())>2) & (df[i].dtypes=='O'):
        print(i+':',len(df[i].value_counts()))
        value_list2.append(i)
value_list2
```

```python
value_list3=[]
for i in df.columns:
    if df[i].dtypes!='O':
        print(i+':',len(df[i].value_counts()))
        value_list3.append(i)
value_list3
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df_encoded1=df[value_list1].apply(LabelEncoder().fit_transform)
df_encoded1.head()
```

```python
df_encoded2=pd.get_dummies(df[value_list2])
df_encoded2
```

```python
new_df=df_encoded1.join([df[value_list3],df_encoded2])
new_df.head()
```

```python
new_df.columns
```
