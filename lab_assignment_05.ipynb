We have collected all the articles from AMIA in the last 30 years. The dataset is avaliable on GitHub: https://raw.githubusercontent.com/unt-iialab/info5502-spring2022/main/datasets/lab_assignment_05_amia_cleaned.csv. 

Please use the tool Yake: https://github.com/LIAAD/yake to extract the top N (you can set different N value to get the best visualization results) keywords from the article titles (all the articles), then visualize how the keywords changed over years in one figure. (Split the data by every five years)

```python
# Your python code here:

#importing libraries
!pip install yake 
import yake
import pandas as pd
import random
import matplotlib.pyplot as plt
#defining the source of data
url = 'https://raw.githubusercontent.com/unt-iialab/info5502-spring2022/main/datasets/lab_assignment_05_amia_cleaned.csv'
#reading the data
df = pd.read_csv(url)
#prinitng out the 100 samples from data
df.head(100)

#Setting the top 2 words for each and every 5 years.
df=pd.read_csv(url)
wordcount = 2
b=0
#logic
while(b<9):
	a=0 + (5*b)
	for y in range(1981+a,1986+a):  
  #setting the list of words
		wlist=" ".join(title for title in df[df['year']==y]
                     ['title'].values) 
  #extracting using yake tool
	extract=yake.KeywordExtractor(lan="en", top=wordcount, features=None) 
	word=extract.extract_keywords(wlist)
 #defining X and Y
	Roc = [y for x, y in word]
	Keywords=[x for x, y in word]
  #Giving titles to the representation
	plt.scatter(Keywords,Roc,label=y) 
	plt.title("Keywords Trend over time")
	b = b+1
#plotting the visualization 
plt.legend(bbox_to_anchor=(1.2, 1.0), loc='upper left')
plt.tick_params(axis='x', labelsize=20)
plt.tick_params(axis='y', labelsize=20)
plt.xticks(rotation=90)
plt.xlabel("Keywords that changed")
plt.ylabel("Rate of change")
plt.show()
b = b+1



```
